```python
import numpy as np
import matplotlib.pyplot as plt
```

# Logarithmic regression

### Prediction

$$
\begin{align*}
y &= a +  \mathbf{w^\intercal}\log{\mathbf{x}} \\
\end{align*}
$$

```python
class LogarithmicRegression:

    def fit(self, X, y):
        X = np.hstack((np.ones((len(X), 1)), X))
        self.weights = np.dot(np.linalg.pinv(np.log(X)), y)
        self.bias = self.weights[0]
        self.weights = self.weights[1:]

    def predict(self, X):
        return self.bias + np.dot(np.log(X), self.weights)
```

```python
def make_logarithmic_regression(n_samples, noise):
    X = np.linspace(0.5, 10, n_samples)
    y = np.log(X) + np.random.normal(0, noise, n_samples)
    return X, y

X, y = make_logarithmic_regression(10, noise=0.1)

plt.scatter(X, y)
plt.show()
```

```python
model = LogarithmicRegression()
model.fit(X.reshape(-1, 1), y)

x = np.linspace(0.1, 15, 100)

plt.scatter(X, y)
plt.plot(x, model.predict(x.reshape(-1, 1)), color='red')
plt.show()
```
